```python
import matplotlib.pyplot as plot
import random
```

```python
def getRandom(dig=4, neg=False):
    start = 0
    if neg:
        start = -(10**dig) + 1
    return random.randint(start, (10**dig) - 1)
```

```python
N = 25000
dig = 4
negative = True
data = []
```

```python
for i in range(N):
    first = getRandom(dig, negative)
    print ("First is " + str(first))
    data.append(first)
    for i in range (3):
        sub = getRandom(dig, negative)
        result = first - sub
        print ("%d - %d = %d" % (result, first, sub))
        data.append(result)
```

```python
plot.hist(data, bins=1000)
```
